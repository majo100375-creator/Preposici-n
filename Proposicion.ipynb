Operadores Lógicos

```python
#Definición de la proposición original
calor = True

#Negacion de la proposición
negacion = not calor

#Imprimir los resultados
print(f"Proposición original, hace calor: {calor}")
print(f"Negación de la proposición original, hace calor: {negacion}")
```

```python
#Ejemplo: Hoy llueve "y" hace frío = conjunción
#Solo será verdadera si ambas son ciertas
#Definimos las dos proposiciones
proposicion1 = True
proposicion2 = False #Debe ser "True" para que al imprimir De True

conjuncion = proposicion1 and proposicion2

print(f"conjuncion: {conjuncion}")
```

```python
#Ejemplo: En una disyuncion basta que una sea verdadera para que todo sea verdadero.
#Ejemplo: Voy al cine "o" me quedo en casa
#Definición de proposiciones

proposicion1 = True
proposicion2 = False

#Disyunción de las dos proposiciones
disyuncion = proposicion1 or proposicion2

print(f"disyuncion: {disyuncion}")
```

```python
#Implicación : "Sí" y "Entonces"
#Ejemplo: Si como chocolate, entonces estaré feliz
como_chocolate = True
estar_feliz = como_chocolate
print(f"Si como chocolate({como_chocolate}), entonces estare feliz {estar_feliz}") #Los parentesis y llaves es solo para diferenciar, no pasa nada si se coloca solo las llaves, es solo visual
```

```python
#Ejemplo de Implicación2
#Si estudio entonces, aprobare el examen
#Solo sera falsa si estrudias pero no apruebas
estudio = False
aprobar_examen = estudio or not estudio #Si estudio: Implica aprobar. pero si no estudio tambiénequivale a aprobar.
print(f"Si estudio {estudio} entonces aprobaré el examen {aprobar_examen}")
```

```python
# Definimos nuestras variables proposicionales (p y q)
llueve = True  # Esto es 'p'
llevo_paraguas = True  # Esto es 'q'

# Representación del condicional (Si p, entonces q)
if llueve:
    llevo_paraguas = True
    print("Como está lloviendo, llevo el paraguas.")
else:
    print("No está lloviendo, así que no es obligatorio llevarlo.")
```

```python
# Definimos las proposiciones
voy_al_cine = True    # p
me_quedo_en_casa = False  # q

# Aplicamos la disyunción (p o q)
if voy_al_cine or me_quedo_en_casa:
    print("El plan es válido: hice alguna de las dos cosas.")
else:
    print("Esto es falso: ¡no hice ninguna de las dos!")
```

```python
# Proposición original (p)
me_gusta_la_pizza = True

# Negación (¬p)
no_es_cierto_que_me_gusta = not me_gusta_la_pizza

print(f"¿Me gusta la pizza?: {me_gusta_la_pizza}")
print(f"¿No es cierto que me gusta?: {no_es_cierto_que_me_gusta}")
```

Vectores - Algebra básica

```python
#Vector de la pelicula Mátrix
matrix = [1, 0, 1, 0, 1] #(Accion, comedia, ciencia ficcion, drama suspenso)
#Vector pelicula El padrino
el_padrino = [0, 0, 0, 1, 0]

similitud = sum([a*b for a, b in zip(matrix, el_padrino)])

print(f"La similitud entre Matrix y El Padrino es {similitud}")
```

```python
#Vector de correo electronico
correo = ["oferta", "gratis", "dinero", "ganar"]

#Vector de palabras comunes en spam

spam_words = ["oferta", "gratis", "dinero" "ganar", "click"]

#Calcular la cantidad de spam

cantidad_spam = sum([1 for word in correo if word in spam_words]) #Recorre cada palabra que hay en la variable correo y verifica si esta en la lista de palabras spam

if cantidad_spam > 2:
  print("El correo es spam")
else:
  print("Este correo es legítimo")
```

```python
genero = ["accion", "comedia", "ciencia ficcion", "drama", "horror" ]

peliculas = { "The Matrix": [1, 0, 1, 0, 0], "Interstellar": [0, 0, 1, 1, 0], "The Hangover": [0, 1, 0, 0, 0], "Die Hard": [1, 0, 0, 0, 0], "Inception": [1, 0, 1, 1, 0], "The Godfather": [0, 0, 0, 1, 0], "Superbad": [0, 1, 0, 0, 0], "Alien": [0, 0, 1, 0, 1], "Friday the 13th": [0, 0, 0, 0, 1], "John Wick": [1, 0, 0, 0, 0] }
def obtener_similitud(nombre1, nombre2):
    v1 = peliculas[nombre1]
    v2 = peliculas[nombre2]
    return sum(a * b for a, b in zip(v1, v2))
pelicula1 = input("Ingrese el nombre de la primera pelicula")
pelicula2 = input("Ingrese el nombre de la segunda pelicula")
if pelicula1 in peliculas and pelicula2 in peliculas:
  similitud_calculada = obtener_similitud(pelicula1, pelicula2)
  print(f"La similitud entre '{pelicula1}' y '{pelicula2}' es: {similitud_calculada}")

  vector1 = peliculas[pelicula1]
  vector2 = peliculas[pelicula2]

  generos_comunes = []
  for i in range(len(genero)):
      if vector1[i] == 1 and vector2[i] == 1:
          generos_comunes.append(genero[i])

  if generos_comunes:
      print(f"Las películas '{pelicula1}' y '{pelicula2}' comparten los siguientes géneros: {', '.join(generos_comunes)}.")
  else:
      print(f"Las películas '{pelicula1}' y '{pelicula2}' no comparten géneros en común.")
else:
  print("Una o ambas películas no se encontraron en la base de datos. Por favor, verifica los nombres.")
  print("Películas disponibles: ", list(peliculas.keys()))
```
